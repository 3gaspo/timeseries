## Data

```python
data_name = "electricity"  # dataset should be accessible at data_path/data_name.csv
users_dim = 1
date_col = None
drop = None
names = None

def load_dataset(data_name="electricity", users_dim=1, date_col=None, drop=None, names=None, verbose=1):
    with warnings.catch_warnings():
        warnings.filterwarnings(
            "ignore",
            message="Could not infer format, so each element will be parsed individually",
            category=UserWarning,
        )
        raw_df = pd.read_csv(data_path + data_name + ".csv", index_col=0, parse_dates=True)

    if verbose:
        print("Raw shape:", raw_df.shape)
        print("Raw index type:", type(raw_df.index))
        print("Raw columns preview:", list(raw_df.columns[:5]))

    data = prepare_pandas_data(raw_df, users_dim=users_dim, date_col=date_col, drop=drop, names=names)

    if verbose:
        print("Prepared shape:", data.shape)
        print("Index type:", type(data.index))
        print("First index:", data.index[0])
        print("Last index:", data.index[-1])
        print("Columns preview:", list(data.columns[:5]))
        display(data.head())

    return data

data = load_dataset(data_name, users_dim=users_dim, date_col=date_col, drop=drop, names=names, verbose=1)
```
